# Association Rule Mining French retail store project(Aditya Chowdhury)

Importing necessary libraries and loading the dataset

```python
import numpy as np
import pandas as pd
#!pip install apyori
from apyori import apriori
store=pd.read_csv('store_data.csv',header=None)
store
```

Converting dataframe to list format

```python
records=store.stack().groupby(level=0).apply(list).tolist()    
```

Implementing apriori algorithm

```python
association_rules=apriori(records,min_support=0.002667,min_confidence=0.2,min_lift=3,min_length=2)
association_results=list(association_rules)
```

Total number of associations

```python
print('Number of associations formed are:',len(association_results))
```

Paired items list with evaluation metrics

```python
for i in association_results:
    pair=i[0]
    items=[x for x in pair]
    print('Rule:',items[0],'->',items[1])
    print('Support:',str(i[1]))
    print('confidence:',str(i[2][0][2]))
    print('Lift:',str(i[2][0][3]))
    print('------------------------------------------------------------------------')
```
